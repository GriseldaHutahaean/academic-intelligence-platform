```python
import sys
import csv

# Gunakan csv module bawaan Python jika pandas tidak tersedia
try:
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    %matplotlib inline
    
    df = pd.read_csv('../Data/dataset.csv')
    print("✓ Dataset loaded with pandas!")
    print(f"  Shape: {df.shape[0]} rows × {df.shape[1]} columns\n")
    
except ImportError as e:
    print("⚠️ pandas not compatible with current Python version")
    print("\n📝 SOLUSI:")
    print("1. Update notebook kernel ke Python 3.12")
    print("2. Atau install dengan: pip install --upgrade pandas numpy")
    print("\n🔄 Restart kernel setelah install")
    df = None
```

Distribusi Target

untuk mengetahui distribusi jumlah mahasiswa berdasarkan status akademiknya, yaitu Graduate, Dropout, dan Enrolled. hasil analisis memberikan gambaran awal mengenai kondisi dataset serta proporsi mahasiswa yang berhasil lulus, masih aktif, atau mengalami dropout.

```python
df['Target'].value_counts()

plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='Target'
)

plt.title('Distribution of Student Outcomes')
plt.xlabel('Student Outcome')
plt.ylabel('Count')

plt.show()
```

Scholarship Holder Analysis

untuk mengetahui hubungan antara status penerima beasiswa dengan hasil akademik mahasiswa. visualisasi ini dapat diamati apakah mahasiswa penerima beasiswa memiliki kecenderungan yang lebih tinggi untuk lulus dibandingkan mahasiswa yang tidak menerima beasiswa.

```python
pd.crosstab(
    df['Scholarship holder'],
    df['Target']
)

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Scholarship holder',
    hue='Target'
)

plt.title('Scholarship Holder vs Student Outcome')
plt.xlabel('Scholarship Holder')
plt.ylabel('Count')

plt.show()
```

Debtor Analysis

untuk melihat pengaruh status tunggakan pembayaran terhadap hasil akademik mahasiswa.membantu mengidentifikasi apakah mahasiswa yang memiliki tunggakan pembayaran lebih berisiko mengalami dropout dibandingkan mahasiswa yang tidak memiliki tunggakan.

```python
pd.crosstab(
    df['Debtor'],
    df['Target']
)

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Debtor',
    hue='Target'
)

plt.title('Debtor Status vs Student Outcome')
plt.xlabel('Debtor Status')
plt.ylabel('Count')

plt.show()
```

Tuition Fees Analysis

untuk mengetahui hubungan antara kelancaran pembayaran biaya kuliah dengan status akademik mahasiswa. dapat diketahui apakah mahasiswa yang rutin membayar biaya kuliah memiliki peluang kelulusan yang lebih tinggi dibandingkan mahasiswa yang mengalami keterlambatan pembayaran.

```python
pd.crosstab(
    df['Tuition fees up to date'],
    df['Target']
)

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Tuition fees up to date',
    hue='Target'
)

plt.title('Tuition Fees Status vs Student Outcome')
plt.xlabel('Tuition Fees Up To Date')
plt.ylabel('Count')

plt.show()
```

Age Analysis

untuk mengevaluasi distribusi usia mahasiswa saat pertama kali masuk perguruan tinggi berdasarkan status akademiknya. dapat diamati apakah terdapat perbedaan karakteristik usia antara mahasiswa yang lulus, masih aktif, maupun yang mengalami dropout.

```python
df.groupby('Target')['Age at enrollment'].describe()

plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Target',
    y='Age at enrollment'
)

plt.title('Age at Enrollment by Student Outcome')
plt.xlabel('Student Outcome')
plt.ylabel('Age')

plt.show()
```

Gender Analysis

untuk mengetahui distribusi hasil akademik mahasiswa berdasarkan gender. membantu mengidentifikasi apakah terdapat perbedaan kecenderungan kelulusan, status aktif, atau dropout antara mahasiswa laki-laki dan perempuan.

```python
pd.crosstab(
    df['Gender'],
    df['Target']
)

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Gender',
    hue='Target'
)

plt.title('Gender vs Student Outcome')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

Correlation Heatmap

untuk mengidentifikasi hubungan antar variabel numerik dalam dataset. dapat diketahui variabel-variabel yang memiliki hubungan kuat sehingga dapat membantu dalam memahami pola data dan menentukan fitur yang berpotensi berpengaruh pada hasil akademik mahasiswa.

```python
corr_matrix = df.corr(numeric_only=True)

corr_matrix

plt.figure(figsize=(15,10))

sns.heatmap(
    corr_matrix,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
